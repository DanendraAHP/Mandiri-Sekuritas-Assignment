# See Data

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/BBCA_historical_data.csv')
df
```

```python
clean_df = df.copy()
clean_df.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']
clean_df = clean_df.iloc[2:]
clean_df
```

```python
clean_df.isnull().sum()
```

```python
clean_df.info()
```

```python
num_cols = ['Close', 'High', 'Low', 'Open', 'Volume']
for col in num_cols:
    clean_df[col] = clean_df[col].astype(float)
```

```python
clean_df.info()
```

```python
clean_df.to_csv('../data/clean_BBCA_historical_data.csv', index=False)
```
